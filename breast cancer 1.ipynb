```python
import pandas as pd
df=pd.read_excel("under treatment.xlsx")
df2=pd.read_excel("death.xlsx")
df3=pd.read_excel("recovered.xlsx")
```

```python
# concat all three dataset
```

```python
df_merged = pd.concat([df, df2, df3], ignore_index=True)
```

```python
df_merged.drop(columns=["treatment_data","id_healthcenter","id_treatment_region","marital_status","marital_length","pregnency_experience"],inplace=True)
```

```python
df_merged.drop(columns=["patient_id","taking_heartMedicine","taking_blood_pressure_medicine"],inplace=True)
```

```python
df_merged['gender']=df_merged['gender'].replace({0:"Female",1:"Male"})
```

```python
df_merged.rename(columns={'age_FirstGivingBirth': 'age_first_giving_birth'}, inplace=True)
```

```python
df_merged.rename(columns={"hereditary_history":"Hereditary","taking_gallbladder_disease_medicine":"gallbladder_disease"}
                   ,inplace=True)
```

```python
# Breast pain vs Cancer 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='breast_pain', hue='Cancer', data=df_merged, palette='pastel')
plt.title('Number of Patients having Breast Pain')
plt.xlabel('Breast Pain (1 = Yes, 0 = No)')
plt.ylabel('Number of Patients')
plt.legend(title='Cancer')
plt.show()
```

```python
# The total number of cancer cases in each age group.
```

```python
sns.barplot(x='age_group', y='Cancer', hue='age_group', data=df_merged, estimator=sum)
plt.title("Cancer Cases by Age Group")
plt.ylabel("Cancer Count")
plt.xlabel("Age Group")
plt.show()
```

```python
#Total Alcohol and smoking by gender
```

```python


sns.barplot(x='gender', y='alcohol', data=df_merged, estimator=sum, color='#a8ddb5', label='Alcohol')
sns.barplot(x='gender', y='smoking', data=df_merged, estimator=sum, color='#c7e9c0', label='Smoking', alpha=0.7)

plt.title("Total Alcohol and Smoking by Gender")
plt.ylabel("Total Count")
plt.legend()
plt.show()

```

```python
# Checking how Age vise cancer spread 
```

```python
avg_thickness = df_merged.groupby('age')['thickness_tumor'].sum().reset_index()


plt.figure(figsize=(10, 5))
sns.lineplot(x='age', y='thickness_tumor', data=avg_thickness, color='green')

plt.title('Average Tumor Thickness by Age')
plt.xlabel('Age')
plt.ylabel('Tumor Thickness (mm)')
plt.grid(True)
plt.show()
```

```python
# Menstrual Age by age
```

```python
female_df2 = df_merged[df_merged['gender'] == 'Female']
birth_by_age = female_df2.groupby('age')['menstrual_age'].sum().sort_index()

birth_by_age.plot(kind='line', color='red')
plt.title("Average Menstrual Age by Age Group (Females Only)")
plt.xlabel("Age")
plt.ylabel("Average Menstrual Age")
plt.grid(True)
plt.show()
```

```python
#Birth control by age-Female
```

```python
import matplotlib.pyplot as plt

female_df2 = df_merged[df_merged['gender'] == 'Female']

birth_control_by_age = female_df2.groupby('age')['birth_control'].sum().sort_index()


birth_control_by_age.plot(kind='line', color='purple')
plt.title("Average Birth Control Usage by Age Group (Females Only)")
plt.xlabel("Age")
plt.ylabel("birth_control")
plt.grid(True)
plt.show()
```
